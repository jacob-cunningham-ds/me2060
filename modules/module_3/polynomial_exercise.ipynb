```julia
function p2(t)
      return -1+5t-4t^2
end
```

```julia
function p2_(t)
      return (t+2)*(t-0)*(t-1)*(-27/(6*(t+2))+1/(2*t))
end
```

```julia
using Plots,LaTeXStrings

t_x = [-2, 0, 1]
t_y = [-27, -1, 0]
t=-2:0.2:2

plot(t,p2.(t),label="Monomial Basis",lc=:black,lw=2,ls=:solid)
plot!(t,p2_.(t),label="Lagrange Basis",lc=:black,lw=2,ls=:dash)
scatter!(t_x, t_y, label="Data", mc=:red, ms=7, ma=0.5)
xlabel!("t")
ylabel!("y")
```
