```julia
function p2(t)
      return -1+5t-4t^2
end
```

```julia
function p2_(t)
      return (t+2)*(t-0)*(t-1)*(-27/(6*(t+2))+1/(2*t))
end
```

```julia
t=-2:0.2:2
using Plots,LaTeXStrings

plot(t,p2.(t),label=L"Version 1 for $p_2(t)$",lc=:black,lw=2,ls=:solid)
plot!(t,p2_.(t),label=L"Version 2 for $p_2(t)$",lc=:green,lw=2,ls=:dash)
xlabel!("t")
ylabel!("y")
```
